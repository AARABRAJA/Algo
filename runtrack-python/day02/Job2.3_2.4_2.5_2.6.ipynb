```python
#Job 2.3
class Board:
    def __init__(self, i, j):
        self.rows = i
        self.columns = j
        self.board = [['O' for _ in range(j)] for _ in range(i)] 

    def play(self, column, color):
        if color not in ['yellow','red']:
            print("Please choose 'red' or 'yellow'.")
            return False
        
        if column < 0 or column >= self.columns:
            print("Please choose a valid column.")
            return False
        
        for row in range(self.rows - 1, -1, -1):
            if self.board[row][column] == 'O':
                self.board[row][column] = color[0].upper()  
                return True
            
        else:
            print("The column chosen is full, please choose another one.")
            return False
    
    def print(self):
        for row in self.board:
            print(' '.join(row))


#Example:
Board_1 = Board(7,6)
Book_1 = Board_1.play(1,"yellow")
Board_1.print()
```

```python
#Job 2.4
def game():
    color_player = 'yellow'
    i, j = 6,5  
    board = Board(i, j)

    while True:
        board.print()
        column = int(input(color_player[0].upper() + "'s turn. Choose a column (0-" + str(j-1) + "): "))
        if board.play(column, color_player):
            if final_winner(board.board, color_player[0].upper()):
                print(f"{color_player.capitalize()} wins the game!")
                break

            if color_player == 'yellow':
                color_player = 'red'
            else:
                color_player = 'yellow'

    
def final_winner(board, player):
    for i in range(len(board)): #Nbr de lignes
        for j in range(len(board[i])): #Nbr de colonnes dans la ligne i+1
            if j + 3 < len(board[i]) and all(board[i][j + k] == player for k in range(4)):
                return True
            if i + 3 < len(board) and all(board[i + k][j] == player for k in range(4)):
                return True
            if i + 3 < len(board) and j + 3 < len(board[i]) and all(board[i + k][j + k] == player for k in range(4)):
                return True
            if i + 3 < len(board) and j - 3 >= 0 and all(board[i + k][j - k] == player for k in range(4)):
                return True
    return False

if __name__ == "__main__":
    game()
```

```python
#Job 2.5
import random

class AI_One:
    def __init__(self):
        self.color = None

    def think(self, board, color):
        self.color = color

        colums_dispo = []
        for col in range(board.columns):
            if board.board[0][col] == 'O':
                colums_dispo.append(col)

        return random.choice(colums_dispo)

class HumanPlayer:
    def __init__(self, name):
        self.name = name

    def play(self, board):
        while True:
            try:
                column = int(input(self.name+"'s turn. Choose a column (0-"+str(board.columns - 1)+ "): "))
                if board.play(column, self.name):
                    return
            except ValueError:
                print("Invalid input. Please enter a valid column number.")

def main():
    i, j = 3,2  #
    board = Board(i, j)
    AI = AI_One()
    human_player = HumanPlayer("red")

    while True:
        board.print()
        human_player.play(board)
        if final_winner(board.board, human_player.name[0].upper()):
            print(f"{human_player.name.capitalize()} wins the game!")
            break

        ai_column = AI.think(board, "yellow")
        if board.play(ai_column, "yellow"):
            print(f"AI plays in column {ai_column}")
            if final_winner(board.board, "Y"):
                print("AI wins the game!")
                break

if __name__ == "__main__":
    main()
```
