```python
#Job 2.0
class Person:
    def __init__(self, name, firstname):
        self.name = name
        self.firstname = firstname

    def SePresenter(self):
        print("My full name is:",self.name,self.firstname)

    # Accesseur pour l'attribut "name"
    def get_nom(self):
        return self.name

    # Mutateur pour l'attribut "lastname"
    def set_nom(self, new_name):
        self.name = new_name

    # Accesseur pour l'attribut "name"
    def get_prenom(self):
        return self.firstname

    # Mutateur pour l'attribut "lastname"
    def set_prenom(self, new_lastname):
        self.firstname = new_lastname

# Exemple classe Personne
Person_1 = Person("Raja", "Aarab")
Person_2 = Person("Sofia", "Betti")

# Utilisation mÃ©thode SePresenter()
Person_1.SePresenter()
Person_2.SePresenter()

# Utilisation accesseurs et mutateurs
print("Before modification :")
print(Person_1.get_nom())
print(Person_1.get_prenom())

Person_1.set_nom("Inaya")
Person_1.set_prenom("Tati")

print("After modification :")
print(Person_1.get_nom())
print(Person_1.get_prenom())
```

```python
#Job 2.1
class Book:
    def __init__(self, title):
        self.title = title

    def print(self):
        print("The title of the book:",self.title)

class Author(Person):
    def __init__(self, name, firstname):
        super().__init__(name, firstname)
        self.work = []

    def listerWork(self):
        print("List of books written by the author:", self.firstname, self.name)
        for book in self.work:
            print(book.title)

    def writeABook(self, title):
        new_book = Book(title)
        self.work.append(new_book)
        return new_book


# Example 
# Author creation
Author_1 = Author("Inaya", "Johnny")

# Books
Book_1 = Author_1.writeABook("Two Year Eight Months and Twenty")
Book_2 = Author_1.writeABook("The Red Sar")

# Displaying the author's books
Author_1.listerWork()

# Title
Book_1.print()
```

```python

#Job 2.2    
class Library:
    def __init__(self, name):
        self.name = name
        self.catalog = {}

    def buyBook(self, author, book_title, quantity):
        for book in author.work:
            if book_title==book.title:
                if book_title not in self.catalog:
                    self.catalog[book_title] = quantity
                else:
                    self.catalog[book_title] += quantity
                return str(quantity) + " copies of '" + book_title + "' added to the catalog."


        return "'" + book_title + "' by " + author.first_name + " " + author.name + " not found in the author's work."


    def inventory(self):
        for book_title, quantity in self.catalog.items():
            print("Library catalog:",book_title,quantity,"copies")
            print("'" + book_title + "': " + str(quantity) + " copies")

    def rent(self, client, book_title):
        if book_title in self.catalog and self.catalog[book_title] > 0:
            if book_title not in client.books:
                client.books[book_title] = 1
            else:
                client.books[book_title] += 1
            self.catalog[book_title] -= 1
            return "'" + book_title + "' rented by " + client.firstname + " " + client.name + "."

        else:
            return "'" + book_title + "' not available for rent."


    def renderBooks(self, client):
        for book_title, quantity in client.books.items():
            if book_title in self.catalog:
                self.catalog[book_title] += quantity
            else:
                self.catalog[book_title] = quantity
        client.books = {}


class Client(Person):
    def __init__(self, name, firstname):
        super().__init__(name, firstname)
        self.books = {}

    def inventory(self):
        print(self.firstname,self.name,"s Book Collection:")
        for book_title, quantity in self.books.items():
            print(book_title,quantity,"copies")


# Authors
Author_1 = Author("Inaya", "Johnny")
Author_2 = Author("Raja", "Aarab")

# Have authors write books
Book_1 = Author_1.writeABook("Deep Learning Essentials")
Book_2 = Author_1.writeABook("Python for Beginners")
Book_3 = Author_2.writeABook("Data Science Essentials")

# Create libraries
Library_1 = Library("Centrale Med Library")
Library_2 = Library("Centrale Marseille Library")

# Have libraries buy books
print(Library_1.buyBook(Author_1, "Deep Learning Essentials", 5))
print(Library_1.buyBook(Author_2, "Data Science Essentials", 3))
print(Library_2.buyBook(Author_1, "Python for Beginners", 2))

# Create customers
customer1 = Client("Johnson", "Jane")
customer2 = Client("Brown", "Bob")

# Have customers rent books
print(Library_1.rent(customer1, "Deep Learning Essentials"))
print(Library_1.rent(customer2, "Deep Learning Essentials"))
print(Library_1.rent(customer2, "Python for Beginners"))
print(Library_2.rent(customer1, "Data Science Essentials"))

# Display inventories
Library_1.inventory()
Library_2.inventory()
customer1.inventory()
customer2.inventory()

# Have customer1 return books
Library_1.renderBooks(customer1)

# Display inventories after returns
Library_1.inventory()
customer1.inventory()

```
